```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as sm
from sklearn.linear_model import LinearRegression 
from scipy import stats
import statsmodels.api as s
data=pd.read_excel("C:/Users/RohanS/Downloads/ict.xlsx")
data
```

```python

data2 = pd.get_dummies(data,columns =['Batting Situation','Result'],dtype=int)

data2
```

```python
def prediction(y2,y3):
    results=[]
    for i in data2['Player'].unique():
        for j in data['Opponent'].unique():
            data3=data2[(data2['Player']==i) & (data2['Opponent']==j)]
            
            
           
            indep=data3['Runs']
            x2=data3['Batting Situation_Batting First']
            x3=data3['Result_lost']
            
            
            model=sm.ols(formula='indep~x2+x3',data=data3).fit()
           
            new_data = pd.DataFrame({
                'x2': [y2],
                'x3': [y3],
               
              
               
            })
            prediction = int(model.predict(new_data).values[0])
            results.append({
                'Player': i,
                'Opponent': j,
                'Prediction': np.abs(prediction)
            })
    
    # Convert list of dictionaries to DataFrame
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    pd.set_option('display.width', None)
    pd.set_option('display.max_colwidth', None)
    results_df = pd.DataFrame(results)
    print(results_df)
prediction(1,0)
    

        
        
    
        
    
```
